### <b>3.2 Apache Sparkflow</b>



<b>Key Components of Spark Workflow</b>
<ol>
    <li><b>Spark Driver</b> - The driver is the central coordinator that runs the user code to create RDDs (Resilient Distributed Datasets), perform transformations and actions, and create the SparkContext. The driver splits the Spark application into tasks and schedules them to run on the executors.</li>
    <li><b>SparkContext</b> - The SparkContext is the heart of a Spark application. It establishes a connection to the Spark execution environment, creates RDDs, accumulators, and broadcast variables, and runs jobs. It acts as the master of the Spark application.</li>
    <li><b>Cluster Manager</b> - The cluster manager is responsible for launching executors and, in some cases, the driver. It is a pluggable component in Spark and can be a standalone cluster manager, Hadoop YARN, or Apache Mesos.</li>
    <li><b>Executors</b> The executors are the distributed workers that run the individual tasks in a Spark job. They provide in-memory storage for RDDs that are cached by the user and return the results to the driver.</li>
</ol>

<b>Spark Job Execution Flow</b>
<ol>
    <li><b>Submitting an Application</b>: The user submits an application using the spark-submit script, which launches the driver program.</li>
    <li><b>Resource Allocation</b>: The driver program requests resources from the cluster manager to launch executors.</li>
    <li><b>Task Execution</b>: Based on the actions and transformations on RDDs, the driver sends work to executors in the form of tasks.</li>
    <li><b>Result Collection</b>: The executors process the tasks and send the results back to the driver through the cluster manager.</li>


## <b>4. Data Staging</b>

A staging area, or landing zone, is an intermediate storage area used for data processing during the extract, transform and load (ETL) process. The data staging area sits between the data source(s) and the data target(s), which are often data warehouses, data marts, or other data repositories.

<ol>
    <li>Extracting data from source systems.</li>
    <li>Loading it into an intermediate storage area.</li>
    <li>Using the staging area during the ETL (Extract, Transform, Load) process.</li>
    <li>Steps involved in the data staging process include data extraction, profiling, cleansing, transformation, validation, integration, temporal storage, and loading.</li>
    <li>Staging data from scraping websites provides a historical snapshot and allows for data discrepancies to be traced back to raw source data.</li>
</ol>
